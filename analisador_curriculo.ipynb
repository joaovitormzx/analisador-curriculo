```python
import re
from collections import Counter

STOPWORDS = {
    "de", "da", "do", "para", "com", "em", "um", "uma",
    "que", "e", "o", "a", "os", "as", "no", "na", "se",
    "por", "mais", "como", "mas", "ao", "ou", "são"
}

def limpar_texto(texto):
    texto = texto.lower()
    texto = re.sub(r'[^\w\s]', '', texto)
    palavras = texto.split()
    palavras = [p for p in palavras if p not in STOPWORDS]
    return Counter(palavras)

def calcular_score(vaga_kw, curriculo_kw):
    em_comum = set(vaga_kw.keys()) & set(curriculo_kw.keys())
    score = len(em_comum) / len(vaga_kw.keys()) * 100
    return round(score, 2), em_comum, set(vaga_kw.keys()) - set(curriculo_kw.keys())
```

```python
 def analisar(vaga_texto, curriculo_texto):
  vaga_kw = limpar_texto(vaga_texto)
  curriculo_kw = limpar_texto(curriculo_texto)
  score, em_comum, faltando = calcular_score(vaga_kw, curriculo_kw)

  print("=" * 50)
  print("📊 RESULTADO DA ANÁLISE")
  print("=" * 50)
  print(f"\n🎯 Score de compatibilidade: {score}%")

  if score >= 70:
    print("✅ Ótimo alinhamento com a vaga!")
  elif score >= 40:
    print("⚠️ Alinhamento médio — pode melhorar")
  else:
    print("❌ Currículo pouco alinhado com essa vaga")

  print(f"\n✔️ Palavras em comum: {len(em_comum)}")
  print(", ".join(sorted(em_comum)))
  print(f"\n❌ Palavras que faltam no currículo: {len(faltando)}")
  print(", ".join(sorted(faltando)))
  print("\n" + "=" * 50)

vaga = """
Estágio em TI. Buscamos estudante com conhecimentos em
Python, Excel, comunicação, organização e suporte técnico.
"""

curriculo = """
Estudante de Gestão em TI. Tenho conhecimentos em Python,
suporte ao usuário e organização de processos.
"""

analisar(vaga, curriculo)
```

```python
def sugerir_melhorias(faltando):
  if not faltando:
    print("🏆 Currículo totalmente alinhado com a vaga!")
    return

  print("💡 SUGESTÕES PARA MELHORAR SEU CURRÍCULO:")
  print("-" * 40)
  for palavra in sorted(faltando):
    print(f"  👉 Adicione '{palavra}' ao seu currículo")
  print("-" * 40)
  print(f"\n📌 Dica: adicionar essas {len(faltando)} palavras")
  print("   pode aumentar seu score significativamente!")

# Chama a função com as palavras que faltam
score, em_comum, faltando = calcular_score(
    limpar_texto(vaga),
    limpar_texto(curriculo)
)

sugerir_melhorias(faltando)
```

```python
def exportar_resultado(score, em_comum, faltando):
  with open("resultado_analise.txt", "w", encoding="utf-8") as f:
    f.write("=" * 50 + "\n")
    f.write("RESULTADO DA ANÁLISE DE CURRÍCULO\n")
    f.write("=" * 50 + "\n\n")
    f.write(f"Score de compatibilidade: {score}%\n\n")
    f.write(f"Palavras em comum ({len(em_comum)}):\n")
    f.write(", ".join(sorted(em_comum)) + "\n\n")
    f.write(f"Palavras que faltam ({len(faltando)}):\n")
    f.write(", ".join(sorted(faltando)) + "\n")
    f.write("\n" + "=" * 50 + "\n")

  print("✅ Resultado salvo em: resultado_analise.txt")

exportar_resultado(score, em_comum, faltando)
```
